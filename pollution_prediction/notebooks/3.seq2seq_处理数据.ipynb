### 2.构造特征数据

```python
pollution_df['date'] = pollution_df.index
```

```python
pollution_df.head()
```

```python
pollution_df['hour'] = pd.to_datetime(pollution_df['date'],format='%Y-%m-%d %H:%M:%S').dt.hour
pollution_df['month'] = pd.to_datetime(pollution_df['date'],format='%Y-%m-%d %H:%M:%S').dt.month
pollution_df['year'] = pd.to_datetime(pollution_df['date'],format='%Y-%m-%d %H:%M:%S').dt.year
```

```python
pollution_df.head()
```

```python
pollution_df['dayOfMonth'] = pd.to_datetime(pollution_df['date'],format='%Y-%m-%d %H:%M:%S').dt.day
```

```python
pollution_df['weekend'] = pd.to_datetime(pollution_df['date'],format='%Y-%m-%d %H:%M:%S').dt.dayofweek
```

```python
pollution_df.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import LabelEncoder


# 整数编码
encoder = LabelEncoder()
pollution_df['wnd_dir_enc'] = encoder.fit_transform(pollution_df['wnd_dir'])

min_max_cols = ['dew','temp','press','wnd_dir_enc','wnd_spd','snow','rain']
# 归一化特征
scaler = MinMaxScaler(feature_range = (0,1))
pollution_df[min_max_cols] = scaler.fit_transform(pollution_df[min_max_cols])
```

```python
pollution_df.head()
```

```python
pollution_df['pollution(t-1)'] = pollution_df['pollution'].shift(1)
pollution_df['pollution(t-2)'] = pollution_df['pollution'].shift(2)
pollution_df['pollution(t-3)'] = pollution_df['pollution'].shift(3)
```

```python
pollution_df.head()
```

```python
pollution_df.dropna(inplace=True)
```
